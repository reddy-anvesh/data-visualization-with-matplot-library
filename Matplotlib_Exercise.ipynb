# Line Plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Creating a figure with subplots
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(10, 3))

# Choosing a color scheme and increasing line width
colors = plt.get_cmap('tab10').colors
linewidth = 2

# Ploting the data on each subplot with thicker lines and custom colors
ax[0].plot(x, y1, linestyle='-', linewidth=linewidth, color=colors[0])
ax[1].plot(x, y2, linestyle='--', linewidth=linewidth, color=colors[1])
ax[2].plot(x, y3, linestyle=':', linewidth=linewidth, color=colors[2])

# Adding gridlines to each subplot
for i in range(3):
    ax[i].grid(linestyle='--', linewidth=0.5, alpha=0.5)

# Adding a legend to each subplot
ax[0].legend(['sine'], loc='upper left')
ax[1].legend(['cosine'], loc='upper left')
ax[2].legend(['tangent'], loc='upper left')

# Increase font size of labels and titles
fontsize = 12
for i in range(3):
    ax[i].set_xlabel('X axis', fontsize=fontsize)
    ax[i].set_ylabel('Y axis', fontsize=fontsize)
    ax[i].set_title(ax[i].get_title(), fontsize=fontsize)

# Adjust subplot spacing
fig.tight_layout()

# Showing the plot
plt.show()
```

# Scatter Plot

```python
# Generating some data
x = np.random.randn(100)
y = x + np.random.randn(100)

# Creating a scatter plot
fig, ax = plt.subplots(figsize=(6, 5))
ax.scatter(x, y, s=50, alpha=0.5, c=x, cmap='cool')

# Adding a trendline to the plot
p = np.polyfit(x, y, 1)
ax.plot(x, np.polyval(p, x), '-', color='red', linewidth=2)

# Adding labels and titles
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('Scatter plot')

# Showing the plot
plt.show()
```

# Bar Plot

```python
# Generating some data
labels = ['A', 'B', 'C', 'D']
values1 = [20, 35, 30, 25]
values2 = [10, 20, 30, 40]

# Creating a bar plot with stacked bars
fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(labels, values1, color='blue', label='Group 1')
ax.bar(labels, values2, bottom=values1, color='green', label='Group 2')

# Adding data labels to the bars
for i, v1, v2 in zip(range(len(labels)), values1, values2):
    ax.text(i, v1/2, str(v1), color='white', ha='center', va='center', fontsize=14)
    ax.text(i, v1+v2/2, str(v2), color='white', ha='center', va='center',fontsize=14)
    
ax.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.5)

ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Bar plot')
ax.legend()

plt.show()
```

# Histogram

```python
# Generating some data
x = np.random.randn(1000)

# Creating a histogram with customized bins
fig, ax = plt.subplots(figsize=(8, 6))
bins = np.linspace(-3, 3, 21)
ax.hist(x, bins=bins, density=True, edgecolor='black', linewidth=1, alpha=0.5, color='green')

# Adding a vertical line at the mean value
mean = np.mean(x)
ax.axvline(mean, linestyle='--', color='red', linewidth=2, label='Mean')

# Adding labels and titles
ax.set_xlabel('X axis')
ax.set_ylabel('Density')
ax.set_title('Histogram')
ax.legend()

# Showing the plot
plt.show()
```

# Box Plot

```python
# Generating some data
np.random.seed(123)
data = np.random.normal(0, 1, size=(100, 3))

# Creating a box plot
fig, ax = plt.subplots(figsize=(8, 6))
bp = ax.boxplot(data, labels=['A', 'B', 'C'], patch_artist=True)

# Customizing the box plot
colors = ['blue', 'green', 'red']
for box, color in zip(bp['boxes'], colors):
    box.set_facecolor(color)
for whisker in bp['whiskers']:
    whisker.set(color='black', linewidth=1.5)
for cap in bp['caps']:
    cap.set(color='black', linewidth=1.5)
for median in bp['medians']:
    median.set(color='black', linewidth=1.5)
for flier in bp['fliers']:
    flier.set(marker='*', markersize=8, alpha=0.5)

# Adding labels and titles
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Box plot')

# Showing the plot
plt.show()
```

# Heatmap

```python
# Generating random data
data = np.random.rand(5, 5)

# Creating heatmap
fig, ax = plt.subplots()
im = ax.imshow(data, cmap='viridis')

# Adding colorbar
cbar = ax.figure.colorbar(im, ax=ax)

# Adding title and axis labels
ax.set_title('Heatmap')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')

# Seting tick labels
ax.set_xticks(np.arange(data.shape[1]))
ax.set_yticks(np.arange(data.shape[0]))
ax.set_xticklabels(['Column {}'.format(i+1) for i in range(data.shape[1])])
ax.set_yticklabels(['Row {}'.format(i+1) for i in range(data.shape[0])])

# Rotating tick labels and seting alignment
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        text = ax.text(j, i, '{:.2f}'.format(data[i, j]),
                       ha="center", va="center", color="w")

plt.show()
```

# Contour plot

```python
x = np.linspace(-3, 3, 256)
y = np.linspace(-3, 3, 256)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contourf(X, Y, Z, cmap='inferno', levels=10, alpha=0.5)  # adding transparency and more contour lines
cbar = plt.colorbar()
cbar.ax.set_ylabel('Data Value')  # adding color bar label
plt.xlabel('X Axis Label')  # adding x axis label
plt.ylabel('Y Axis Label')  # addig y axis label
plt.title('Contour Plot of Sinusoidal Surface')  # adding a descriptive title
plt.show()
```

# Polar plot

```python
theta = np.linspace(0, 2*np.pi, 100)
r1 = np.cos(3*theta)
r2 = np.sin(2*theta)

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

ax.plot(theta, r1, color='red', label='Cosine')
ax.plot(theta, r2, color='blue', label='Sine')
ax.set_title('Polar Plot of Cosine and Sine')
ax.legend(loc='upper right', fontsize='small')  # adding legend
ax.set_rticks(np.arange(-1, 1.1, 0.5))  # adding radial ticks
ax.grid(True)  # adding grid lines
plt.show()
```

# 3D plot

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.8)  # adding transparency and adjust colors
ax.set_xlabel('X Axis Label')  # adding x axis label
ax.set_ylabel('Y Axis Label')  # adding y axis label
ax.set_zlabel('Z Axis Label')  # adding z axis label
ax.set_title('3D Surface Plot of Sinusoidal Function')  # adding a descriptive title
plt.show()
```
